```python
import sqlite3 as lite
```

```python
con = lite.connect('weather.db')
```

```python
cur = con.cursor()
```

```python
with con:
    cur.execute("create table city_weather (time int, city text, maxtemp numeric) ")
```

```python
sql = "insert into city_weather (time,city,maxtemp) values (?,?,?)"
```

```python
cities.keys()
```

```python
with con:
    for c in cities.keys():
        date = start_date
        for i in range(30):    
            r = requests.get('https://api.forecast.io/forecast/892c4551f3da38b34b8f5d2a4aaeaa26/%s,%s' % (cities[c],date.strftime('%s')))
            maxt = r.json()['daily']['data'][0]['temperatureMax']
            cur.execute(sql,(int(date.strftime('%s')),c,maxt))
            date = date + datetime.timedelta(days=1)
```

```python
import pandas as pd
```

```python
df = pd.read_sql_query("select * from city_weather",con)
```

```python
df
```

```python
df1 = df[df.city == 'Atlanta']
df2 = df[df.city == 'Austin']
df3 = df[df.city == 'Boston']
df4 = df[df.city == 'Chicago']
df5 = df[df.city == 'Providence']
```

```python
dff = df1
```

```python
dff = dff.merge(df2,on='time')
dff = dff.merge(df3,on='time')
dff = dff.merge(df4,on='time')
dff = dff.merge(df5,on='time')
```
